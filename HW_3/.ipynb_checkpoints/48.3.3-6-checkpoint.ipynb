# Task 5

$\lbrace n\Omega_n^2 > u_{1 - \alpha} \rbrace$

```python
from tqdm import tqdm
```

```python
n = [25, 100, 500 ,2000]
alphas = [0.01, 0.05, 0.1]
```

```python
def omega_n_squared(n, sample):
    '''Можно и руками посчитать, выйдет такая сумма
        
    '''
    cdfs = sps.norm.cdf(sorted(sample))
    to_count = [(2 * (i + 1) - 1) / n * np.log(cdfs[i]) + np.log(1 - cdfs[n - 1 - i])  for i in range(n)]
    return -n - np.sum(to_count)
```

```python
stats = []
for size in tqdm(n):
    to_push = []
    for j in range(1000):
        cur_sample = sps.norm.rvs(size=size)
        to_push.append(omega_n_squared(size, cur_sample))
    stats.append(to_push)
```

```python
plt.hist(stats[0])
plt.show()
```

```python
plt.hist(stats[1])
plt.show()
```

```python
plt.hist(stats[2])
plt.show()
```

```python
plt.hist(stats[3])
plt.show()
```

```python
import statsmodels.api as sm
```

```python
sm.qqplot(np.array(stats[3]))
```

```python
sps.probplot(stats[3], plot=plt)
plt.show()
```

```python
sps.shapiro(stats[3])
```

```python
quantiles = []
for alpha in tqdm(alphas):
    cur_quant = []
    for stat in stats:
        cur_quant.append(np.percentile(stat, 1 - alpha))
        
    quantiles.append(cur_quant)
```

```python
quantiles
```

H0 : P = N (0, 1) против альтерна- тивы H1 : P = T (10), где T (10) – распределение Стьюдента с 10 степенями свободы.

Найдем теперь мощность критерия(1 - вероятность ошибки второго рода). То есть сгенерируем выборки из альтернативы

```python
studenta = sps.t(10)
answers = []
ksmirnovs = []
for ind in tqdm(range(len(n))):
    to_push = []
    to_ksmirnovs = []
    for second_ind in range(len(alphas)):
        vals = []
        ks = []
        for j in range(100):
            samplee = studenta.rvs(size=n[ind])
            vals.append(omega_n_squared(n[ind], samplee))
            ks.append(sps.kstest(samplee, 'norm')[1])
        
        to_push.append(sum(np.array(vals) > quantiles[second_ind][ind]))
        to_ksmirnovs.append(sum(np.array(ks) > alphas[second_ind]))
            
    answers.append(to_push)
    ksmirnovs.append(to_ksmirnovs)
```

```python
alphas
```

```python
answers
```

```python
ksmirnovs
```

Видно, что Андерсен-Дарлинг отработал лучше, чего мы и ожидали!

Проверить гипотезу $H_0 : P \in \mathcal{P} = {P_\theta , \theta \in (0; 0.5)} $ против альтернативы $H_1 : P \notin \mathcal{P}$ где

$$P_\theta(\lbrace 0 \rbrace) = -\frac{\sqrt{\theta}}{ln(\theta)}, P_\theta(\lbrace 1 \rbrace) = \frac{1}{2}e^{-\theta}$$

```python
mus = np.array([[67,33,0], [44,12,44]])
n = np.sum(mus, axis=1)
```

```python
import scipy
```

```python
def fun(theta):
    return (mushki[0] * (0.5 * np.log(theta) - np.log(-np.log(theta))) + \
           mushki[1] * (-np.log(2) - theta) + \
           (mushki[2] * np.log(1 + np.sqrt(theta) / np.log(theta) - 0.5 * np.exp(-theta)) if mushki[2] != 0 else 0 )
           
           ) * -1.
```

Найдя ОМП для $\theta$ Теперь запишем статистику $\chi^2$, подставив в нее омп 

$\chi^2(\hat{\theta}) = \sum\limits_{i = 1}^{n}\frac{(\mu_i - np_i(\hat{\theta}))^2}{np_i(\hat{\theta})}$

```python
pdfs = [
lambda theta: - np.sqrt(theta) / np.log(theta),
lambda theta: 0.5 * np.exp(-theta),
lambda theta: (1 + np.sqrt(theta) / np.log(theta) - 0.5 * np.exp(-theta))
]
```
