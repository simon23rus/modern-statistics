```python
import statsmodels.api as sm
```

Мы не можем отвергнуть гипотезу нестационарности на уровне значимости 0.05, попробуем преобразовать ряд, например, сделав дифференцирование. Вроде дисперсия у нас не сильно меняется со временем, поэтому просто дифференцирования должно хватить

```python
print("Критерий Дики-Фуллера: p=%f" % sm.tsa.stattools.adfuller(dollar.diff1[1:])[1]) # (1:) так как 0 элемент - NAN
```

Круто, теперь гипотеца о нестационарности отвергается на уровне 0.05 и мы можем приступить к подбору параметров модели ARIMA. Для этого нам понадобятся графики частичной и обычной Автокорреляционной функции 

```python
plt.figure(figsize=(15,8))
ax = plt.subplot(211)
sm.graphics.tsa.plot_acf(dollar.diff1[1:].values.squeeze(), lags=50, ax=ax)
plt.show()
plt.figure(figsize=(15,8))
ax = plt.subplot(212)
sm.graphics.tsa.plot_pacf(dollar.diff1[1:].values.squeeze(), lags=50, ax=ax)
plt.show()
```

На графике автокорреляции с лагом 10 видно значимое отличие от нуля автокорреляции, тогда возьмем q = 10

А на частной автокорреляции можно взять p = 7

Приступим к подбору оптимального параметров, для этого пройдемся по сетке и воспользуемся информативными критериями AIC(BIC)

```python
ps = range(0, 8)
d=1
qs = range(0, 11)
```

```python
from itertools import product
parameters = product(ps, qs)
parameters_list = list(parameters)
len(parameters_list)
```

```python
from tqdm import tqdm
```

```python
import statsmodels.tsa.arima_model
```

```python
import warnings
```

```python
%%time
results = []
best_aic = float("inf")
warnings.filterwarnings('ignore')
d=1
for param in tqdm(parameters_list):
    #try except нужен, потому что на некоторых наборах параметров модель не обучается
    try:
        model=sm.tsa.ARIMA(dollar.price, order=(param[0], d, param[1])).fit(disp=-1)
    #выводим параметры, на которых модель не обучается и переходим к следующему набору
    except ValueError:
        print('wrong parameters:', param)
        continue
    aic = model.aic
    #сохраняем лучшую модель, aic, параметры
    if aic < best_aic:
        best_model = model
        best_aic = aic
        best_param = param
    results.append([param, model.aic])
    
warnings.filterwarnings('default')
```

```python
results
```

AIC Меньше -> лучше

```python
result_table = pd.DataFrame(results)
result_table.columns = ['parameters', 'aic']
print(result_table.sort_values(by = 'aic', ascending=True).head())
```

```python
print(best_model.summary())
```

Посмотрим, насколько же хорошую модель мы построили

```python
plt.figure(figsize=(15,8))
plt.subplot(211)
best_model.resid[13:].plot()
plt.ylabel(u'Residuals')

ax = plt.subplot(212)
sm.graphics.tsa.plot_acf(best_model.resid[13:].values.squeeze(), lags=50, ax=ax)

print("Критерий Стьюдента: p=%f" % sps.ttest_1samp(best_model.resid[13:], 0)[1])
print("Критерий Дики-Фуллера: p=%f" % sm.tsa.stattools.adfuller(best_model.resid[13:])[1])
```

На уровне значимости 0.05 мы не можем отвергнуть гипотезу о том, что среднее отлично от 0, а гипотеза нестационарности отвергается

Так же видно по коррелограмме, что остатки некоррелируемые

# Task 6 $R(s,t) = e^{-|t - s|}$

Он центрированный, тогда мы все знаем про процесс и можем генерить тренировочнуб выборку размера n, что значит просто сгенерить один элемент из n-мерного нормального распределения с данной нам ковариационной функцией

Возьмем $t_{train} = range(n)$

```python
from sklearn.metrics import pairwise_distances
```

```python
n = 1000
t_train = np.array(range(n))
sigma = pairwise_distances(t_train.reshape(n,1), t_train.reshape(n,1), metric='l1')
```

```python
sigma = np.exp(-sigma)
```

```python
train = sps.multivariate_normal(cov=sigma).rvs()
```

```python
plt.figure(figsize=(15,7))
plt.plot(range(n), train, color=col_gen.get_color())
plt.show()
```

А теперь мы хотим смоделировать наш процесс на будущее, используя модели Авторегрессии AR(1) и AR(2)

```python
from statsmodels.tsa.ar_model import AR
```

```python
ar_1 = AR(train)
ar_2 = AR(train)
```

```python
fitted_1 = ar_1.fit(1)
fitted_2 = ar_2.fit(2)
```
